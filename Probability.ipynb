```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("student-mat.csv")
df.head(3)
```

```python
len(df)
```

#### Add a boolean column called grade_A noting if a student achieved 80% or higher as a final score.

```python
df['grade_A'] = np.where(df['G3']*5 >= 80, 1, 0)
```

#### Make another boolean column called high_abscences with a value of 1 if a student misses 10 or more classes.

```python
df['high_absences'] = np.where(df['absences'] >= 10, 1, 0)
```

#### Add one more column to make building the pivot table easier

```python
df['count'] = 1
```

```python
df
```

```python
df = df[['grade_A', 'high_absences', 'count']]
df.head()
```

```python
pd.pivot_table(df,
              values = 'count',
              index = ['grade_A'],
              columns = ['high_absences'],
              aggfunc = np.size,
              fill_value = 0
              )
```

#### In our case, P(A) is the probability of a grade of 80% or greater. P(B) is the probability of missing 10 or more classes. P(A|B) is the probability of a 80%+ grade, given missing 10 or more classes.

Calculations of parts:

###### P(A)
= (35 + 5)/(35 + 5 + 277 + 78) <br>
= 0.10126582278481013

###### P(B)
