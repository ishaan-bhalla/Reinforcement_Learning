#### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
np.random.seed(1)
```

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)
```

```python
x
```

#### Positive Correlation with some noise

```python
y = x + np.random.normal(0, 10, 1000)
y
```

```python
np.corrcoef(x, y)
```

```python
%matplotlib inline
matplotlib.style.use("ggplot")
plt.scatter(x, y)
plt.show()
```

#### Negative Correlation with some noise

```python
y = 100 - x + np.random.normal(0, 10, 1000)
y
```

```python
np.corrcoef(x, y)
```

```python
%matplotlib inline
matplotlib.style.use("ggplot")
plt.scatter(x, y)
plt.show()
```

```python
df = pd.DataFrame({"a" : np.random.randint(0, 50, 1000)})
df["b"] = df["a"] + np.random.normal(0, 10, 1000) # positively coorelated with "a"
df["c"] = 100 - df["a"] + np.random.normal(0, 5, 1000) # negatively coorelated with "a"
df["d"] = np.random.randint(0, 50, 1000) # not coorelated with "a"
```

```python
df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize = (6, 6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```
